```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px

df_generator_1 = pd.read_csv("Plant_1_Generation_Data.csv", parse_dates=['DATE_TIME'])
df_generator_2 = pd.read_csv("Plant_2_Generation_Data.csv", parse_dates=['DATE_TIME'])

df_concat = pd.concat([df_generator_1, df_generator_2], ignore_index=True)
df_filtered = df_concat.dropna()

start_date = '2020-05-15 00:00'
end_date = '2020-05-22 00:00'

data_mask = (df_filtered['DATE_TIME'] >= start_date) & (df_filtered['DATE_TIME'] < end_date)
specific_data = df_filtered[data_mask & (df_filtered['SOURCE_KEY'] == 'McdE0feGgRqW7Ca')]
```

```python
df_with_avg_for_power_of_gen = pd.DataFrame(df_filtered[data_mask].groupby('SOURCE_KEY')['AC_POWER'].mean())
df_with_avg_for_power_of_gen.rename(columns={'AC_POWER':'AVG_AC_POWER'}, inplace = True)

fig = px.line(specific_data, x="DATE_TIME", y="AC_POWER")

for i in range(df_with_avg_for_power_of_gen.size):
    fig.add_shape(type='line', y0=df_with_avg_for_power_of_gen['AVG_AC_POWER'][i],
                  y1=df_with_avg_for_power_of_gen['AVG_AC_POWER'][i],
                  x0=start_date,
                  x1=end_date,
                  line=dict(color='red'))
fig.update_layout(title = "Wykres zależności AC_POWER od DATE_TIME wraz ze średnimi AC_POWER dla wszystkich generatorów")
fig.show()
```

```python
df_merged = pd.merge(df_filtered[data_mask], df_with_avg_for_power_of_gen, on='SOURCE_KEY', how='right')
df_AC_below_avg = df_merged[ df_merged['AC_POWER'] < 0.8*df_merged['AVG_AC_POWER'] ]
df_AC_below_avg.SOURCE_KEY.value_counts()
```
